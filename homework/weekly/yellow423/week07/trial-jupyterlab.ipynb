        // 平滑滚动
        document.querySelectorAll('a[href^="#"]').forEach(anchor => {
            anchor.addEventListener('click', function (e) {
                e.preventDefault();
                document.querySelector(this.getAttribute('href')).scrollIntoView({
                    behavior: 'smooth'
                });
            });
        });

        // 图片懒加载
        document.addEventListener("DOMContentLoaded", function() {
            const lazyImages = [].slice.call(document.querySelectorAll("img"));
            
            if ("IntersectionObserver" in window) {
                let lazyImageObserver = new IntersectionObserver(function(entries, observer) {
                    entries.forEach(function(entry) {
                        if (entry.isIntersecting) {
                            let lazyImage = entry.target;
                            lazyImage.classList.add("fade-in");
                            lazyImageObserver.unobserve(lazyImage);
                        }
                    });
                });
                
                lazyImages.forEach(function(lazyImage) {
                    lazyImageObserver.observe(lazyImage);
                });
            }
        });
    </script>

牛顿-莱布尼茨公式：
\[
\int_{a}^{b} f(x) \, dx = F(b) - F(a)
\]
其中 \( F \) 是 \( f \) 的一个原函数。

多元函数的泰勒展开式：
\[
f(\mathbf{x}) = f(\mathbf{a}) + \nabla f(\mathbf{a})^T (\mathbf{x} - \mathbf{a}) + \frac{1}{2} (\mathbf{x} - \mathbf{a})^T H_f(\mathbf{a}) (\mathbf{x} - \mathbf{a}) + \cdots
\]
其中 \( H_f(\mathbf{a}) \) 是 \( f \) 在点 \( \mathbf{a} \) 处的 Hessian 矩阵。

\subsection{概率论与统计学}
多元正态分布的概率密度函数：
\[
f(\mathbf{x}; \boldsymbol{\mu}, \boldsymbol{\Sigma}) = \frac{1}{(2\pi)^{n/2} |\boldsymbol{\Sigma}|^{1/2}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{x} - \boldsymbol{\mu}) \right)
\]
其中 \( \mathbf{x} \in \mathbb{R}^n \)，\( \boldsymbol{\mu} \) 是均值向量，\( \boldsymbol{\Sigma} \) 是协方差矩阵。

\subsection{量子力学}
薛定谔方程：
\[
i\hbar \frac{\partial}{\partial t} \Psi(\mathbf{r}, t) = \left[ -\frac{\hbar^2}{2m} \nabla^2 + V(\mathbf{r}, t) \right] \Psi(\mathbf{r}, t)
\]

\subsection{机器学习}
神经网络的反向传播算法中的梯度计算：
\[
\frac{\partial \mathcal{L}}{\partial W_{ij}^{(l)}} = \sum_{k} \delta_k^{(l)} a_j^{(l-1)}
\]
其中 \( \delta_k^{(l)} \) 是误差项，\( a_j^{(l-1)} \) 是前一层的激活值。

\subsection{特殊函数与恒等式}
贝塞尔函数的积分表示：
\[
J_n(x) = \frac{1}{\pi} \int_{0}^{\pi} \cos(n\theta - x\sin\theta) \, d\theta
\]

黎曼 ζ 函数与素数的关系（欧拉乘积公式）：
\[
\zeta(s) = \prod_{p \text{ prime}} \frac{1}{1 - p^{-s}} = \sum_{n=1}^{\infty} \frac{1}{n^s} \quad (\Re(s) > 1)
\]
