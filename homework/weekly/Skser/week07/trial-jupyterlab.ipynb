\section*{示例数学公式}

以下是几个典型的数学公式示例：

\subsection*{1. 一元二次方程求根公式}
对于形如 $ ax^2 + bx + c = 0 $ 的一元二次方程，其解为：
$$
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$

\subsection*{2. 泰勒展开式}
函数 $ f(x) $ 在点 $ a $ 附近的泰勒展开式为：
$$
f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!}(x-a)^n
$$

\subsection*{3. 概率密度函数（正态分布）}
正态分布 $ N(\mu, \sigma^2) $ 的概率密度函数为：
$$
f(x) = \frac{1}{\sqrt{2\pi}\sigma} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

\subsection*{4. 矩阵运算}
设有两个矩阵 $ A $ 和 $ B $，它们的乘积定义如下：
$$
A = 
\begin{bmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{bmatrix}, \quad
B = 
\begin{bmatrix}
b_{11} & b_{12} \\
b_{21} & b_{22}
\end{bmatrix}, \quad
AB = 
\begin{bmatrix}
a_{11}b_{11} + a_{12}b_{21} & a_{11}b_{12} + a_{12}b_{22} \\
a_{21}b_{11} + a_{22}b_{21} & a_{21}b_{12} + a_{22}b_{22}
\end{bmatrix}
$$

\subsection*{5. 多重积分}
三维空间中的体积积分可以表示为：
$$
V = \iiint_{D} f(x,y,z) \, dx\,dy\,dz
$$

\subsection*{6. 极限与导数}
极限定义：
$$
\lim_{x \to a} f(x) = L
$$
导数定义：
$$
f'(x) = \frac{d}{dx}f(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}
$$
